```python

"""
*************************************
*************************************
******      OBJECTIVE - 1      ******
****** MOST INVOLVED CANDIDATE ******
*************************************
*************************************
"""
from IPython.display import Image
from collections import Counter

candidates_list = ["BIDEN", "BOOKER", "BUTTIGIEG", "GABBARD", "HARRIS", "KLOBUCHAR", "SANDERS", "STEYER", "WARREN", "YANG"]
candidates_occurence_list = []
input_file = open("2019_debate5_input_files\\2019_debate5_transcript.txt", "r", encoding="utf8")
for line in input_file:
    if '\n' != line:
        if line.split(':')[0] in candidates_list:
            candidates_occurence_list.append(line.split(':')[0])
candidates_occurence_counter=Counter(candidates_occurence_list)
most_involved_candidate=str(candidates_occurence_counter.most_common(1)).strip('[]()').split(',')[0].replace('\'','')
Image(filename='2019_debate5_images\\'+str(most_involved_candidate).lower()+'.jpg', height=200, width=200) 
```

```python
"""
**************************************
**************************************
******       OBJECTIVE - 2      ******
***** MOST FREQUENTLY USED WORDS *****
**************************************
**************************************
"""

from collections import Counter
import re
from nltk.corpus import stopwords
import nltk
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%matplotlib inline

with open('2019_debate5_input_files\\2019_debate5_transcript.txt', 'r', encoding="utf8") as file:
    data = file.read().replace('\n', ' ')
word_list = data.split()
cleaned_word_list = []
new_cleaned_word_list = []
word_cloud_words = ' '
stop_words = stopwords.words('english')

for word in word_list:
    cleaned_word_list.append(re.sub(r'[\,\.\[\]\-\!\?\:\)\(]','',word.lower()))

for cleaned_word in cleaned_word_list:
    if cleaned_word in stop_words or len(cleaned_word) < 4:
        cleaned_word_list.remove(cleaned_word)
    else:
        new_cleaned_word_list.append(cleaned_word)

for words in new_cleaned_word_list:
    word_cloud_words = word_cloud_words + words + ' '
wordcloud = WordCloud(background_color ='white', min_font_size = 12).generate(word_cloud_words)
plt.figure(figsize = (5, 10), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show()

most_frequent_words_list = Counter(new_cleaned_word_list).most_common(10)
for frequent_word in most_frequent_words_list:
    print(frequent_word)
```

```python
"""
***********************************************
***********************************************
******           OBJECTIVE - 3           ******
***** MOST WORDS USED ANSWERING QUESTIONS *****
***********************************************
***********************************************
"""

from IPython.display import Image
import operator
candidates_list = ["BIDEN", "BOOKER", "BUTTIGIEG", "GABBARD", "HARRIS", "KLOBUCHAR", "SANDERS", "STEYER", "WARREN", "YANG"]
most_number_of_words_used={}
current_candidate="Not Applicable"
old_candidate="Not Applicable"

input_file = open("2019_debate5_input_files\\2019_debate5_transcript.txt", "r", encoding="utf8")
for line in input_file:
    if '\n' != line:
        current_candidate=line.split(':')[0]
        if current_candidate in candidates_list:
            old_candidate=current_candidate
            if current_candidate in most_number_of_words_used:
                most_number_of_words_used[current_candidate] = int(most_number_of_words_used[current_candidate]) + int(len(line.split(" ")))
            else:
                most_number_of_words_used[current_candidate] = str(len(line.split(" ")))
        else:
            if old_candidate in most_number_of_words_used:
                most_number_of_words_used[old_candidate] = int(most_number_of_words_used[old_candidate]) + int(len(line.split(" ")))
            else:
                most_number_of_words_used[old_candidate] = str(len(line.split(" ")))

maxnum=-1
for i in most_number_of_words_used:
    if most_number_of_words_used[i] > maxnum:
        most_words_used_candidate=i
        maxnum=most_number_of_words_used[i]
Image(filename='2019_debate5_images\\'+str(most_words_used_candidate).lower()+'.jpg', height=200, width=200) 
```

```python
"""
***************************************************
***************************************************
******             OBJECTIVE - 5             ******
***** MOST ADDRESSED CANDIDATE BY JOURNALISTS *****
***************************************************
***************************************************
"""

from IPython.display import Image
import operator
journalists_list = ["MADDOW", "MITCHELL", "PARKER", "WELKER"]
candidates_list = ["BIDEN", "BOOKER", "BUTTIGIEG", "GABBARD", "HARRIS", "KLOBUCHAR", "SANDERS", "STEYER", "WARREN", "YANG"]
most_addressed_candidate={}
input_file = open("2019_debate5_input_files\\2019_debate5_transcript.txt", "r", encoding="utf8")
for line in input_file:
    if '\n' != line:
        current_journalist=line.split(':')[0]
        if current_journalist in journalists_list:
            line_splitter=line.split(' ')
            for word in line_splitter:
                if word.upper() in candidates_list:
                    if word.upper() in most_addressed_candidate:
                        most_addressed_candidate[word.upper()] = int(most_addressed_candidate[word.upper()]) + 1
                    else:
                        most_addressed_candidate[word.upper()] = 1
maxnum=-1
for i in most_addressed_candidate:
    if most_addressed_candidate[i] > maxnum:
        result=i
        maxnum=most_addressed_candidate[i]
Image(filename='2019_debate5_images\\'+str(result).lower()+'.jpg', height=200, width=200) 
```
